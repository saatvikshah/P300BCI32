##Tabulated Accuracy and Visualization

```python
from pandas import read_csv,DataFrame as df
import pandas
from ggplot import *
%matplotlib inline
```

```python
# Lets check the result for a single LDA classifier
results = read_csv("results_LDA.csv")
print results
```

```python
# Lets Visualize it
dfplot = pandas.melt(results[['Index','8','10','12','13','15']],id_vars=['Index'])
plot = ggplot(aes(x='Index',y='value', group='variable',colour='variable'),data=dfplot) +\
    geom_line() + geom_point() + xlab("Feature Index #") + ylab("Accuracy(%)")
print plot
```

```python
# Let's check results for a Bagged Ensemble of 500 LDA Classifiers
results = read_csv("results_ens.csv")
print results
```

```python
# Lets Visualize it
dfplot = pandas.melt(results[['Index','8','10','12','13','15']],id_vars=['Index'])
plot = ggplot(aes(x='Index',y='value', group='variable',colour='variable'),data=dfplot) +\
    geom_line() + geom_point() + xlab("Feature Index #") + ylab("Accuracy(%)")
print plot
```

We can clearly see that ensembling helps boost the accuracy for certain features, when using it with lower sequences of training data. Thus a choice for **Ensembling Linear Classifiers** ameliorates prediction results
